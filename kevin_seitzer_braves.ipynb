```python
from pybaseball import team_batting
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
batting_data = team_batting(2007, 2024, ind=1)
batting_data
```

```python
atlanta_batting_data = batting_data[batting_data["Team"] == "ATL"]
```

```python
atlanta_batting_data
```

```python
atlanta_batting_data.sort_values(by=['Season'])
```

```python
print(atlanta_batting_data.columns.tolist())
```

```python
fig, ax = plt.subplots()

ax.scatter(atlanta_batting_data['Season'], atlanta_batting_data['OPS'], color='blue')

plt.title('Braves OPS by Year')
ax.set_xlabel('Year')
ax.set_ylabel('OPS')

plt.xticks([2008, 2010, 2012, 2014, 2016, 2018, 2020, 2022, 2024])

ax.axvline(x = 2014.5, ymin=.05, ymax=.9, color='black', linestyle='--', label='Kevin Seitzer joins Braves')

arrow1 = mpatches.FancyArrowPatch((2007, .79), (2013.5, .73), mutation_scale=20, linewidth=.01, color='red')
arrow2 = mpatches.FancyArrowPatch((2015, .7), (2019.5, .84), mutation_scale=20, linewidth=.01, color='green')

ax.add_patch(arrow1)
ax.add_patch(arrow2)

plt.legend(loc='upper left')

plt.savefig('Braves_OPS_by_Year.png')

plt.show()
```
