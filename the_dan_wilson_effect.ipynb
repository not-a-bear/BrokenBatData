```python
from pybaseball import playerid_reverse_lookup
```

```python
dan_all_batting = pd.read_csv('dan_all_batting.csv')
```

```python
scott_all_batting = pd.read_csv('scotty_all_batting.csv')
```

```python
dan_all_batting
```

```python
dan_batter_list = []
for batter_id in dan_all_batting['batter']:
    dan_batter_list.append(batter_id)
dan_batter_set = set(dan_batter_list)
dan_batter_set
```

```python
scott_batter_list = []
for batter_id in scott_all_batting['batter']:
    scott_batter_list.append(batter_id)
scott_batter_set = set(scott_batter_list)
scott_batter_set
```

```python
scott_batters = []
scott_batters = playerid_reverse_lookup(scott_batter_set, key_type='mlbam')
scott_batters
```

```python
dan_batters = []
dan_batters = playerid_reverse_lookup(dan_batter_set, key_type='mlbam')
dan_batters
```

```python
common_batter_set = dan_batter_set.intersection(scott_batter_set)
common_batter_set
```

```python
common_batters = playerid_reverse_lookup(common_batter_set, key_type='mlbam')
common_batters
```

```python
common_batters[['name_last', 'name_first', 'key_mlbam']]
```

```python
scott_pitches_faced = set()
for pitch in scott_all_batting['pitch_type']:
    scott_pitches_faced.add(pitch)
scott_pitches_faced
```

```python
dan_pitches_faced = set()
for pitch in dan_all_batting['pitch_type']:
    dan_pitches_faced.add(pitch)
dan_pitches_faced
```

```python
common_pitches_faced = scott_pitches_faced.intersection(dan_pitches_faced)
common_pitches_faced
```

```python
scott_pitches_faced.difference(dan_pitches_faced)
```

```python
dan_pitches_faced.difference(scott_pitches_faced)
```

```python
batter_pitch_dict_scott = {}
for batter in common_batter_set: 
    for pitch in common_pitches_faced:
        key = (batter, pitch)
        temp_data = scott_all_batting[(scott_all_batting['batter'] == batter) & (scott_all_batting['pitch_type'] == pitch)]
        value = temp_data['delta_run_exp'].sum()
        batter_pitch_dict_scott[key] = value
batter_pitch_dict_scott
```

```python
deltas_array_scott = []
batter_deltas_list = []
for batter in common_batter_set:
    for pitch in common_pitches_faced:        
        batter_deltas_list.append(batter_pitch_dict_scott[(batter, pitch)])
    deltas_array_scott.append(batter_deltas_list)
    batter_deltas_list = []
deltas_array_scott
```

```python
batters_vs_pitches_scott = pd.DataFrame(deltas_array_scott, index = list(common_batter_set), columns = list(common_pitches_faced))
batters_vs_pitches_scott
```

```python
batter_pitch_dict_dan = {}
for batter in common_batter_set: 
    for pitch in common_pitches_faced:
        key = (batter, pitch)
        temp_data = dan_all_batting[(dan_all_batting['batter'] == batter) & (dan_all_batting['pitch_type'] == pitch)]
        value = temp_data['delta_run_exp'].sum()
        batter_pitch_dict_dan[key] = value
batter_pitch_dict_dan
```

```python
deltas_array_dan = []
batter_deltas_list = []
for batter in common_batter_set:
    for pitch in common_pitches_faced:        
        batter_deltas_list.append(batter_pitch_dict_dan[(batter, pitch)])
    deltas_array_dan.append(batter_deltas_list)
    batter_deltas_list = []
deltas_array_dan
```

```python
batters_vs_pitches_dan = pd.DataFrame(deltas_array_dan, index = list(common_batter_set), columns = list(common_pitches_faced))
batters_vs_pitches_dan
```

```python
batters_vs_pitches_difference = batters_vs_pitches_dan - batters_vs_pitches_scott
batters_vs_pitches_difference
```

```python
plt.figure()

outs = ['field_out', 'grounded_into_double_play', 'force_out']
results = ['Strikeout', 'Walk', 'Single', 'Out', 'Hit by Pitch']
cols_list = ['red', 'lime', 'blue', 'yellow', 'black']
cols = np.where(julio_vs_CH_scott_abridged['events'] == 'strikeout','red',
               np.where(julio_vs_CH_scott_abridged['events'] == 'walk','lime',
             np.where(julio_vs_CH_scott_abridged['events'] == 'single', 'blue',
                     np.where(julio_vs_CH_scott_abridged['events'].isin(outs), 'yellow', 'black'))))

temp_patch = patches.Patch()
patch_list = []
for result, col in zip(results, cols_list):
    temp_patch = patches.Patch(color=col, label=result)
    patch_list.append(temp_patch)

fig, ax = plt.subplots()

zonebox = patches.Rectangle((-.7, 1.5), 1.4, 2, linewidth = 1, edgecolor = 'black', facecolor = 'none')

plt.scatter(julio_vs_CH_scott_abridged['plate_x'], julio_vs_CH_scott_abridged['plate_z'], s=209, c=cols, edgecolor='dimgray', linewidth=1, alpha=.75)

plt.title('Julio vs. Changeups Leading to Events Under Scott')

plt.xlabel('Pitch Position x (ft)')

plt.ylabel('Pitch Position y (ft)')

ax.add_patch(zonebox)

plt.grid(False)

ax.legend(handles=patch_list, loc = 'center right', bbox_to_anchor=(1.3, .5))

ax.set_xlim([-2, 2])

ax.set_ylim([0, 4.5])

plt.savefig('julio_vs_CH_scott.png', bbox_inches='tight')

plt.show()
```
